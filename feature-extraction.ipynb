```python
import pandas as pd
import pickle
import numpy as np
```

```python
df = pickle.load(open('dataset_level2.pkl','rb'))
```

```python
df

# batting_team
# bowling team
# city
# current score
# ball left
# wickets left
# current rr
# last five
```

```python
df.isnull().sum()
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df['city'] = cities
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['venue'],inplace=True)
```

```python
df
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
#df['current_score'] = df.groupby('match_id').cumsum()['runs']
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python



df['player_dismissed'] = df['player_dismissed'].apply(lambda x: x if x == "0" else "1")
df['player_dismissed'] = df['player_dismissed'].astype('int')

# Check if there are any non-integer values in 'player_dismissed'
non_integer_values = df[~df['player_dismissed'].astype(str).str.isdigit()]['player_dismissed']
if not non_integer_values.empty:
    print(f"Non-integer values found in 'player_dismissed': {non_integer_values}")

# Now the 'player_dismissed' column is of numeric type, you can proceed with the cumulative sum
wickets = df.groupby('match_id')['player_dismissed'].cumsum().values
df['wickets_left'] = 10 - wickets
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
# Convert 'runs' column to numeric type, coercing errors to NaN
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')

# Group by 'match_id'
groups = df.groupby('match_id')

# Get unique match IDs
match_ids = df['match_id'].unique()

# Initialize the list to store the rolling sums
last_five = []

# Iterate over match IDs
for match_id in match_ids:
    # Extract the group for the current match_id
    current_group = groups.get_group(match_id)

    # Apply rolling sum to the 'runs' column (assuming you want a 30-game rolling sum)
    rolling_sum = current_group['runs'].rolling(window=30, min_periods=1).sum()

    # Extend the last_five list with the rolling sum values
    last_five.extend(rolling_sum.tolist())
```

```python
df['last_five'] = last_five
```

```python
#df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','']]
df['last_five']
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
